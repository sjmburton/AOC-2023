```python
data = open("2023-01.txt", "r")
data = data.read()

data = data.split("\n")
print(data)


        

            
```

```python
# Part 1
number_list = []
for item in data:
    digit_list = []
    for char in item:
        if char.isdigit():
            digit_list.append(char) 
    if len (digit_list) == 0:
        number_list.append(0)
    if len (digit_list) == 1:
        number_list.append(int(str(digit_list[0])+str(digit_list[0])))
    if len (digit_list) > 1:
        number_list.append(int(str(digit_list[0])+str(digit_list[-1])))


output = sum(number_list)
print(output)
```
